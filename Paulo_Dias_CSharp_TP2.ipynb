```python
try
{
    Age age = new();

    while (true)
    {

        Console.WriteLine("Por favor digite sua data de nascimento (Formato DD/MM/AAAA): ");

        string birthday = Console.ReadLine();

        if (DateTime.TryParse(birthday, out DateTime CBirthday))
        {
            age.Ages(CBirthday);
            break;
        }
        else
        {
            Console.WriteLine("Data fornecida Inválida, tente novamente!");
        }
    }
}
    catch (Exception ex)
{
    Console.WriteLine($"Ocorreu um erro: {ex.Message}");
}

                
```

##### Essa estrutura permite o usuário inserir a data de nascimento, e o sistema se encarregará de garantir a inserção antes de direcionar ao método Ages();

##### Código correspondente a estrutura Ages();

```python
public void Ages(DateTime birthday)
{
    DateTime now = DateTime.Now;

    int yearsOfAge = now.Year - birthday.Year;
    int difMonth = now.Month - birthday.Month;
    int difDays = now.Day - birthday.Day;


    if (now.Month < birthday.Month)
    {
        yearsOfAge--;
    }
    else if (now.Month == birthday.Month && now.Day < birthday.Day)
    {
        yearsOfAge--;
    }

    if (difMonth < 0)
    {
        difMonth += 12;
    }

    if (difDays < 0)
    {
        DateTime dateLastMonth = now.AddMonths(-1);
        difDays += DateTime.DaysInMonth(dateLastMonth.Year, dateLastMonth.Month);
        difMonth--;
    }

    Console.WriteLine($"Você possui {yearsOfAge} anos, {difMonth} meses e {difDays} dias.");

}
```

##### Essa estrutura recebe a data de aniversário como a variável "birthday" e faz a sistinção entre a data de hoje e a data de nascimento. à partir desse ponto temos a tratativa temporal para identificar o intervalo, sendo que se o mês atual for menor que o mês de aniversário, indica que ainda não se completou o ano, e fazemos essa redução, caso seja falso, ele identifica se ambos os mêses são iguais E se o dia atual é menor que o dia de aniversário, sendo o caso, é reduzido um ano da diferença pois indica que ainda não se completou o ano. à seguir, caso a diferença de mêses seja negativa, indica que o mês de aniversário é superior ao mês atual, sendo assim adicionamos 12 (referente aos 12 mêses do ano) ao valor negativo, chegando ao valor correto da diferença de mêses. Por fim, caso a diferença de dias seja negativa, indica que o mês de aniversário é superior ao mês atual, sendo que não foi completado o ciclo. Assim crio uma variável "dateLastMonth" apontando para um mês antes do mês atual, à seguir atribuo os dias do mês anterior juntamente com a diferença negativa à variável "difDays" através do método "DaysInMonth()", assim corrigimos a conta de dias e reduzimos um mês da diferença de meses. ### Exercício 2 - Dias até o Próximo Aniversário

#### Código correspondente:

```python
try
{

    Age age = new();

    Console.WriteLine("Qual seu primeiro nome? ");
    string name = Console.ReadLine();
    Console.WriteLine();

    while (true)
    {
        Console.WriteLine("Por favor entre sua data de nascimento (formato DD/MM/AAAA): ");
        string birthday = Console.ReadLine();


        if (DateTime.TryParse(birthday, out DateTime CBirthday))
        {
            age.Ages(CBirthday, name);
            break;
        }
        else
        {
            Console.WriteLine("Por favor entre uma data válida!");
        }
    }
}
catch (Exception ex)
{
    Console.WriteLine($"Ocorreu um erro: {ex.Message}");
}
```

##### Neste código, pedimos o nome de usuário, após dentro da estrutura loop, solicitamos a data de nascimento do usuário, primeiro como uma string, e após efetivando o devido tratamento com a estrutura TryParse. Estando tudo certo, chamo o método Ages() para realizar o devido tratamento.

##### Código correspondente ao método Ages() (Sobrecarga):

```python
public void Ages(DateTime birthday, string name)
{
    DateTime now = DateTime.Now;

    DateTime birthdayThisYear = new DateTime(now.Year, birthday.Month, birthday.Day);

    if (birthdayThisYear > now)
    {
        TimeSpan remainingDays = birthdayThisYear - now;
        Console.WriteLine($" Seu aniversário será ainda este ano e faltam {remainingDays.Days} dias para chegar, {name}!");
    }
    else
    {
        DateTime birthdayNextYear = new DateTime(now.Year + 1, birthday.Month, birthday.Day);
        TimeSpan remainingDays = birthdayNextYear - now;
        Console.WriteLine($" Seu aniversário será no próximo ano e faltam {remainingDays.Days} dias para chegar, {name}!");
    }

}
```

##### Nesta estrutura, após receber a data de nascimento através da variável "birthday", crio uma nova variável chamada "birthdayThisYear" com a mesma data e mês que a variável "birthday", apenas alterando o ano para o ano atual (2025). Após, caso o aniversário neste ano seja maior que a data atual, faço a diferença entre as datas através do tipo "TimeSpan" e exibo a informação de dias que faltam para o anivesário chegar. Caso a data de aniversário seja menor que a data atual, indicando que o aniversário do usuário este ano já aconteceu, crio uma variável chamada "birthdayNextYear" para indicar o aniversário no próximo ano, com o mesmo princípio que a "birthdayThisYear" mas adicionando +1 a variável ano. Após é realizado o mesmo procedimento através do TimeSpan, e exibindo o intervalo de dias que faltam para chegar a data de aniversário.

### Exercício 3 - Diferença entre duas datas

##### Código correspondente:

```python
 try
{
    Age age = new();

    while (true)
    {
        Console.WriteLine("Por favor digite uma data aleatória (Formato DD/MM/AAAA): ");
        string firstDate = Console.ReadLine();

        if (DateTime.TryParse(firstDate, out DateTime CFirstDate))
        {
            Console.WriteLine("Por favor digite outra data aleatória(Formato DD / MM / AAAA): ");
            string secondDate = Console.ReadLine();

            while (true)
            {
                if (DateTime.TryParse(secondDate, out DateTime CSecondDate))
                {
                    age.DifDate(CFirstDate, CSecondDate);

                    break;
                }
            }
            break;
        }
        else
        {
            Console.WriteLine("Por favor entre uma data válida!");
        }
    }
}
catch (Exception ex)
{

    Console.WriteLine($"Ocorreu um erro: {ex.Message}");

}
```

##### Nesta estrutura inicio solicitando uma data aleatória, e tentando efetivar a conversão de string para data, sendo bem sucedido, peço mais uma data aleatória ao usuário, e tento novamente a conversão da segunda data. Sendo bem sucedido, entro no método DifDate() dentro de Age.

##### Código correspondente a DifDate():

```python
public void DifDate(DateTime CFirstDate, DateTime CSecondDate)
{

    int difYear = (int)(CSecondDate.Year - CFirstDate.Year);
    int difMonth = (int)(CSecondDate.Month - CFirstDate.Month);
    int difDay = (int)(CSecondDate.Day - CFirstDate.Day);

    TimeSpan totalDifInDays = CSecondDate - CFirstDate;
    int CTotalDifInDays = Math.Abs(totalDifInDays.Days);

    if (difDay < 0)
    {
        difMonth--;
        difDay += DateTime.DaysInMonth(CSecondDate.Year, CSecondDate.Month);
    }
    if (difMonth < 0)
    { 
        difMonth += 12;
        difYear--;
    }

    if (CFirstDate < CSecondDate)
    {
        Console.WriteLine();
        Console.WriteLine("Primeira data é menor que a segunda");
        Console.WriteLine($"Diferença: {difYear} anos, {difMonth} meses, {difDay} dias.");
        Console.WriteLine($"Total de dias: {CTotalDifInDays} dias.");
    }
    else if (CFirstDate > CSecondDate)
    {
        Console.WriteLine();
        Console.WriteLine("A primeira data é maior que a segunda");
        Console.WriteLine($"Diferença: {difYear} anos, {difMonth} meses, {difDay} dias.");
        Console.WriteLine($"Total de dias: {CTotalDifInDays} dias.");
    }
    else
    {
        Console.WriteLine();
        Console.WriteLine("As duas datas são iguais.");
    }

}
```

##### Semelhante ao código anterior, realizo a diferença entre as os dois intervalos, subtraindo o primeiro date do segundo, separando as variáveis por dia, mês e ano. Também, efetivo a verfificação da diferença total em dias usando o tipo TimeSpan, e após uso Math.Abs() para garantir que o valor correspondente ao total da diferença em dias seja positivo. Igualmente ao exercício anterior, caso a diferença localizada em dias seja negativa, indica que o dia da segunda variável é inferior ao da primeira variável, dessa forma reduzimos um mês da variável de diferença de meses e realizamos o tratamento da variável diferença de dias, adicionando os dias correspondentes ao mês anterior da segunda data. O processo se repete para a variável de diferença de mês caso seja negativa. Após imprimo as informações no console indicando a diferença de dias, mêses, e anos, e a diferença de dias totais. Considerando que as duas datas podem ser iguais.

### Exercício 6 - Cálculo de IMC

##### Código Correspondente:

```python
try
{
    while (true)
    {
        Console.WriteLine("Digite seu peso: ");
        string weight = Console.ReadLine();
        Console.WriteLine();

        if (int.TryParse(weight, NumberStyles.AllowDecimalPoint, CultureInfo.InvariantCulture, out int CWeight))
        {
            while (true)
            {
                Console.WriteLine("Digite a sua altura: ");
                string height = Console.ReadLine();
                Console.WriteLine();

                if (double.TryParse(height, NumberStyles.AllowDecimalPoint, CultureInfo.InvariantCulture, out double CHeight))
                {

                    part2 BMICalculation = new();
                    BMICalculation.BMICalculation(CWeight, CHeight);
                    break;

                }
                else
                {
                    Console.WriteLine("Dado inválido, por favor digite um número válido");
                }

            }
            break;
        }
        else
        {
            Console.WriteLine("Dado inválido, por favor digite um número válido");
        }

    }
}
catch (Exception ex)
{
    Console.WriteLine($"Ocorreu um erro: {ex.Message}");
}
```

##### Nesta execução, inicio com a solicitação da altura da pessoa, tentando a conversão para o tipo int, tomando o cuidado de garantir que seja visível duas casas decimais na variável, utilizando NumberStyles.AllowDecimalPoint e CultureInfo.InvariantCulture. E o mesmo para o peso, mas o tornando um tipo double. Essa abordagem garante o cálculo preciso caso o usuário digite alguma pontuação na altura e no peso. Após, chamo o método BMICalculation presente dentro de part2.

##### Código correspondente a BMICalculation():

```python
public void BMICalculation(int CWeight, double CHeight)
{ 
    double BMI = CWeight / Math.Pow(CHeight, 2);

    string result = BMI switch
    {
        < 18.5 => "Abaixo do peso",
        >= 18.5 and <= 24.9 => "Peso normal",
        >= 25 and <= 29.9 => "Sobrepeso",
        >= 30 and <= 34.9 => "Obesidade Grau I",
        >= 35 and <= 39.9 => "Obesidade Grau II",
        >= 40 => "Obesidade Mórbida",
        _ => "Fora do Intervalo",
    };

    Console.WriteLine($" IMC: {BMI:F2} \n " +
                      $"Indica: {result}");
}
```

##### Aqui efetivo a conta do IMC, armazenando o resultado na variável BMI tipo double. Através de uma estrutura switch, efetivo a verificação do valor de BMI com a mensagem correspondente.

##### Código correspondente a WageDeduction():

```python
public void WageDeductions(double wage)
{

    // INSS Logic Site: https://www.contabilizei.com.br/contabilidade-online/desconto-inss/

    double INSS;

    if (wage <= 1518.00)
    {
        INSS = wage * 0.075;
    }
    else if (wage >= 1518.01 && wage <= 2793.88)
    {
        INSS = wage * 0.09 - 22.77;
    }
    else if (wage >= 2793.89 && wage <= 4190.84)
    {
        INSS = wage * 0.12 - 106.59;
    }
    else if (wage >= 4190.85 && wage <= 8157.41)
    {
        INSS = wage * 0.14 - 190.40;
    }
    else
    {
        INSS = 0.00;
    }


    // IRPF Logic Site: https://www.contabilizei.com.br/contabilidade-online/desconto-inss/

    double IRPF;

    if (wage <= 2259.20)
    {
        IRPF = 0.00;
    }
    else if (wage >= 2259.21 && wage <= 2826.65)
    {
        IRPF = wage * 0.075 - 169.44;
    }
    else if(wage >= 2826.66 && wage <= 3751.05)
    {
        IRPF = wage * 0.15 - 381.44;
    }
    else if (wage >= 3751.06 && wage <= 4664.68)
    {
        IRPF = wage * 0.225 - 662.77;
    }
    else
    {
        IRPF = wage * 0.275 - 896;
    }

    double liquidWage = wage - INSS - IRPF;

    Console.WriteLine($"Salário Bruto: R$ {wage:F2} \n" +
                      $"Desconto INSS: -R$ {INSS:F2} \n" +
                      $"Desconto IRPF: -R$ {IRPF:F2} \n" +
                      $"Salário Líquido: R$ {liquidWage:F2} \n");

}
```

##### Neste código, é verificado a faixa salarial do usuário, e armazenando os descontos respectivos nas variáveis INSS e IRPF. Ao final, é exibido o valor do salário bruto, os respectivos descontos e o salário líquido.

### Exercício 12 - Jogo de Adivinhação

##### Código correspondente:

```python
try
{
    while(true)
    {
        System.Random random = new Random();
        int value = random.Next(1, 101);

        int count = 5;

        while (count > 0)
        {
            Console.WriteLine();
            Console.WriteLine($"Faça um chute entre 1 e 100 (Você tem {count} tentativas): ");
            string guess = Console.ReadLine();
            Console.WriteLine();

            if (int.TryParse(guess, out int CGuess))
            {
                if (CGuess == value)
                {
                    Console.WriteLine($"Muito bem, você acertou, o valor misterioso era {value}");
                    break;
                }
                else if (CGuess < value)
                {
                    Console.WriteLine($"Ainda não, o valor misterioso está pelo menos à {(value / CGuess)} valores acima");
                    count--;
                }
                else
                {
                    Console.WriteLine($"Não não, o valor misterioso está pelo menos à {(CGuess / value)} valores abaixo");
                    count--;
                }
            }
            else
            {
                Console.WriteLine("Valor digitado inválido!");
            }
            if (count == 0)
            {
                Console.WriteLine($"Infelizmente você perdeu. O valor misterioso era {value}.");
            }
        }
        
        break;
    }
}
catch ( Exception ex )
{
    Console.WriteLine($"Ocorreu um erro: {ex.Message}");
}
```

##### Esse código usa uma estrutura bem simples, sendo solicitado um "chute" do usuário, sendo um valor entre 1 e 100 com 5 tentativas para o acerto. À cada incersão de valor, o programa verifica a veracidade e tenta converter para tipo int de um tipo string. Após e comparado se o chute está de acordo com o valor gerado pelo método Random(). Caso as variáveis sejam iguais, o programa anuncia a igualdade e encerra o jogo, caso o chute seja menor que o valor, é operado uma divisão do valor randomico pelo chute e apontado aproximadamente em qual direção o jogador precisa mirar (Sendo à baixo neste caso) dizendo que o valor está à pelo menos X casas naquela direção. O mesmo se aplica caso o chute seja maior que o valor randomico. E por fim, caso o jogador tenha extinguido as 5 tentativas, o jogo encerra com uma mensagem de "Infelizmente você perdeu"
